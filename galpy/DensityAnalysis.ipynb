```python
import SCFAnalysis.Plot as plot
from SCFAnalysis.Constants import *
from SCFAnalysis.func import *
import numpy as nu
from galpy.potential import SCFPotential, plotDensities, plotPotentials
from galpy import potential

import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
%matplotlib inline  
```

# Added bump $10^{-2}$ at $(n, l) = (20, 9)$

```python
folder = "SCF_N30_L15_l9n20_1e-2"
loc = "SCFAnalysis/Orbits/{0}".format(folder)
Acos = nu.load(loc + "/Acos.npy")
plot.plotAxiCoeffs(Acos, .1)


scf = SCFPotential(Acos=Acos, a=200./8,normalize=.35)
plotDensities(scf)


radius = nu.array([0.3, 0.6,1.,1.5, 2])[:,None]
theta = nu.linspace(0, nu.pi, 100)[None, :]
R = radius*nu.cos(theta)
z = (radius**2 - R**2)**.5

plt.figure()
for i in range(len(radius)):
    plt.plot(theta[0,:], scf.dens(R[i],z[i]), label="radius = {0}".format(radius[i,0]))
    
plt.xlabel(r"$\theta (rads)$")
plt.ylabel("Density")
plt.legend()
```
