```python
import numpy as nu
import matplotlib.pyplot as plt
from galpy.potential import KuzminDiskPotential
%matplotlib inline  
```

## Kuzmin Disk Potential
$$\phi(R,z) = -\frac{amp}{\sqrt{R^2 + (a + |z|)^2}}$$

```python
a = 1.
amp = 1.
kd = KuzminDiskPotential(amp, a)
```

```python
kd.plot()
```

##Forces
$$-\frac{\partial \phi}{\partial R} = -\frac{amp}{\sqrt[3]{R^2 + (a + |z|)^2}} R$$

$$-\frac{\partial \phi}{\partial z} = -\frac{amp}{\sqrt[3]{R^2 + (a + |z|)^2}} \frac{z}{|z|} (a + |z|)$$

## Rotation Curves

Setting the b parameter of the Miyamoto Nagai Potential to zero generalizes it to a Kuzmin Disk Potential. Setting b=0, its rotation curve should overlap with Kuzmin Disk rotation curve.

```python
from galpy.potential import MiyamotoNagaiPotential, plotRotcurve
mn = MiyamotoNagaiPotential(amp, a, 0.)
mn.plotRotcurve(overplot=True)
kd.plotRotcurve(overplot=True)
```

##Orbit

```python
from galpy.orbit import Orbit
o= Orbit(vxvv=[1.,0.1,1.1,0.,0.1])
```

```python
import numpy as nu
ts= nu.linspace(0,100,10000)
o.integrate(ts,kd,method='odeint')
```

Plotting the orbit

```python
o.plot()
```

Is energy conserved?

```python
o.plotE(normed=True)
```

As we see, the energy changes by a magnitude of $10^{-5}$.
